```python
import networkx as nx

# 1. Inisiasi Knowledge Base (Database Skala Besar)
G = nx.DiGraph()

data_sejarah = [
    # Tokoh Utama & Kemerdekaan
    ("Soekarno", "merumuskan", "Pancasila"),
    ("Soekarno", "adalah", "Presiden Pertama RI"),
    ("Presiden Pertama RI", "dijuluki", "Bapak Proklamator"),
    ("Soekarno", "lahir di", "Surabaya"),
    ("Surabaya", "berada di", "Indonesia"),
    ("Soekarno", "membacakan", "Teks Proklamasi"),
    ("Soekarno", "didampingi oleh", "Mohammad Hatta"),
    ("Mohammad Hatta", "menyusun", "Teks Proklamasi"),
    ("Mohammad Hatta", "adalah", "Wakil Presiden Pertama RI"),
    ("Wakil Presiden Pertama RI", "dijuluki", "Bapak Koperasi"),
    ("Teks Proklamasi", "menandakan", "Kemerdekaan Indonesia"),
    ("Kemerdekaan Indonesia", "pada tanggal", "17 Agustus 1945"),
    ("Teks Proklamasi", "dibacakan di", "Jakarta"),
    ("Jakarta", "berada di", "Indonesia"),
    ("Teks Proklamasi", "disusun di", "Rumah Laksamana Maeda"),
    ("Rumah Laksamana Maeda", "berlokasi di", "Jakarta"),

    # Tokoh Tambahan & Peristiwa
    ("Sayuti Melik", "mengetik", "Teks Proklamasi"),
    ("Fatmawati", "menjahit", "Bendera Pusaka"),
    ("Fatmawati", "adalah", "Istri Soekarno"),
    ("Bendera Pusaka", "dikibarkan pada", "Kemerdekaan Indonesia"),
    ("Golongan Muda", "menculik", "Soekarno"),
    ("Golongan Muda", "menculik", "Mohammad Hatta"),
    ("Golongan Muda", "membawa ke", "Rengasdengklok"),
    ("Rengasdengklok", "adalah", "Tempat Pengamanan Tokoh"),
    ("Laksamana Maeda", "adalah", "Perwira Angkatan Laut Jepang"),
    ("Laksamana Maeda", "membantu", "Penyusunan Proklamasi")
]

# Membangun edge (garis penghubung)
for subjek, relasi, objek in data_sejarah:
    G.add_edge(subjek, objek, label=relasi)

# 2. Mesin Inferensi Cerdas (Dengan NLP Final)
def jawab_pertanyaan(pertanyaan):
    q = pertanyaan.lower()
    jawaban_ditemukan = False
    print("-" * 65)

    # NLP 1: ALIAS & SINONIM
    sinonim_entitas = {
        "proklamasi": "teks proklamasi",
        "hatta": "mohammad hatta",
        "karno": "soekarno",
        "kemerdekaan": "kemerdekaan indonesia",
        "merdeka": "kemerdekaan indonesia",
        "bendera": "bendera pusaka",
        "maeda": "laksamana maeda",
        "sayuti": "sayuti melik",
        "istri soekarno": "fatmawati",
        "wakil presiden pertama": "wakil presiden pertama ri",
        "presiden pertama": "presiden pertama ri"
    }
    for kata, ganti in sinonim_entitas.items():
        if kata in q and ganti not in q:
            q = q.replace(kata, ganti)

    # NLP 2: LOGIKA KATA TANYA
    if "dimana" in q or "di mana" in q:
        if "baca" in q: q += " dibacakan di"
        elif "susun" in q or "buat" in q: q += " disusun di"
        elif "lahir" in q: q += " lahir di"
        else: q += " berada di berlokasi di"

    if "siapa" in q:
        if "susun" in q or "buat" in q: q += " menyusun"
        elif "baca" in q: q += " membacakan"
        elif "ketik" in q: q += " mengetik"
        elif "jahit" in q: q += " menjahit"
        elif "culik" in q or "bawa" in q: q += " menculik membawa ke"
        elif "damping" in q or "teman" in q: q += " didampingi oleh"
        elif "rumus" in q: q += " merumuskan"
        else: q += " adalah"

    if "kapan" in q or "tanggal" in q:
        q += " pada tanggal dikibarkan pada"

    if "juluk" in q or "gelar" in q:
        q += " dijuluki"

    fakta_spesifik = []

    # 3. PENCARIAN DI KNOWLEDGE BASE
    for subjek, objek, data_relasi in G.edges(data=True):
        s = subjek.lower()
        o = objek.lower()
        r = data_relasi['label'].lower()

        akar_relasi = r
        if "baca" in r: akar_relasi = "baca"
        if "susun" in r: akar_relasi = "susun"
        if "ketik" in r: akar_relasi = "ketik"
        if "jahit" in r: akar_relasi = "jahit"
        if "culik" in r: akar_relasi = "culik"
        if "rumus" in r: akar_relasi = "rumus"
        if "damping" in r: akar_relasi = "damping"
        if "juluk" in r: akar_relasi = "juluk"

        subjek_atau_objek_cocok = (s in q) or (o in q)

        # --- PERBAIKAN: Mencegah Entitas Tumpang Tindih (Overlapping) ---
        if (s == "presiden pertama ri" or o == "presiden pertama ri") and "wakil" in q:
            subjek_atau_objek_cocok = False
        # -----------------------------------------------------------------

        relasi_cocok = (r in q) or (akar_relasi in q)

        if subjek_atau_objek_cocok and relasi_cocok:
            fakta_spesifik.append(f"✅ Jawaban: {subjek} {data_relasi['label']} {objek}")
            jawaban_ditemukan = True

    # Cetak hasil pasti
    if jawaban_ditemukan:
        for fakta in set(fakta_spesifik):
            print(fakta)
    else:
        # NLP 3: FALLBACK (Jaring pengaman)
        fakta_umum = []
        for node in G.nodes():
            if node.lower() in q:
                for subjek, objek, data_relasi in G.edges(data=True):
                    if subjek == node or objek == node:
                        fakta_umum.append(f"-> {subjek} {data_relasi['label']} {objek}")

        if fakta_umum:
            print(f"ℹ️ Saya kurang mengerti maksud pertanyaannya, tapi ini yang saya tahu:")
            for fakta in set(fakta_umum):
                print(fakta)
            jawaban_ditemukan = True

    if not jawaban_ditemukan:
        print("❌ Maaf, data tersebut tidak ada di Knowledge Base kelompok kami.")
    print("-" * 65)


# ==========================================
# 4. ANTARMUKA CHATBOT (JALANKAN DI SINI)
# ==========================================
print("\n" + "=" * 65)
print(" 🚀 AI SEJARAH KELOMPOK 5 (VERSI FINAL & TERLENGKAP) AKTIF")
print("=" * 65)
print("Silakan uji AI kami dengan pertanyaan seperti:")
print("- Siapa wakil presiden pertama?")
print("- Apa julukan presiden pertama?")
print("- Siapa yang mengetik proklamasi?")
print("- Dimana teks proklamasi dibacakan?")
print("- Rengasdengklok itu apa?")

while True:
    tanya = input("\nTanya AI (ketik 'exit' untuk berhenti): ")
    if tanya.lower() in ['keluar', 'exit', 'stop', 'quit']:
        print("Sistem AI dimatikan. Terima kasih telah mencoba!")
        break
    jawab_pertanyaan(tanya)
```
